```python
#Behzad Khadim
#SP20-BCS-019

#all sentences in the assignment 
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import math

sentence1 = 'sunshine state enjoy sunshine'
sentence2 = 'brown fox jump high, brown fox run'
sentence3 = 'sunshine state fox run fast'

corpus = [sentence1,sentence2,sentence3]
corpus
```

```python

vectorizer = CountVectorizer()

data = vectorizer.fit_transform(corpus)
headings = vectorizer.get_feature_names()
array = data.toarray()
print(headings)
print(data.toarray())
```

```python
indexes = ['S1','S2','S3']

table = pd.DataFrame(array,columns = [headings],index= indexes)

table
```

Term Frequency for the given sentences

```python
def termf(sentence, wordbag):
    term_frequency_dictionary = {}
    length_of_wordbag = len(wordbag)

    for word, count in sentence.items():
        term_frequency_dictionary[word] = count / float(length_of_wordbag)

    return term_frequency_dictionary
```

```python
dicS1 = table.iloc[0].to_dict()
bags1 = sentence1.split(" ")
tfs1 = termf(dicS1, bags1)

dicS2 = table.iloc[1].to_dict()
bags2 = sentence2.split(" ")
tfs2 = termf(dicS2, bags2)

dicS3 = table.iloc[2].to_dict()
bags3 = sentence3.split(" ")
tfs3 = termf(dicS3, bags3)


term_frequency = pd.DataFrame([tfs1,tfs2,tfs3], columns = [headings])
term_frequency
```

invserse document frequency for the given sentences

```python
def invserse(sentence):
    idfDict = {}
    N = len(sentence)
    
    idfDict = dict.fromkeys(sentence[0].keys(), 0)     # Idf 
    for doc in sentence:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict
```

```python
iout = invserse([dicS1, dicS2, dicS3])

inverse = pd.DataFrame([iout],columns =[headings])
inverse
```

TF-IDF for the given data

```python
def termandinverse(tfsentence, idfs):
    tfidf = {}
    for word, val in tfsentence.items():
        tfidf[word] = val*idfs[word]
    return tfidf

new1 = termandinverse(tfs1,iout)
new2 = termandinverse(tfs2,iout)
new3 = termandinverse(tfs3,iout)

tfidf = pd.DataFrame([new1,new2,new3], columns = [headings])
tfidf
```
